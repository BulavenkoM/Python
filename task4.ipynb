# Списки, кортежі, словники, множини та виключення в Python

## Опис

Ваше завдання — створити програму для мандрівника у часі, яка буде використовувати різні типи даних та обробку виключень у Python.

## Особливості

- **Списки**: Вам потрібно створити список, кожен елемент якого буде кортежем із роком та подією в цьому році.
- **Кортежі**: Кожен кортеж повинен містити рік та опис події.
- **Словники**: Створіть словник, у якому ключі будуть роками, а значення — подіями.
- **Множини**: Створіть множину років, які ви хочете відвідати.
- **Виключення**: Ваш код повинен коректно обробляти ситуацію, коли зазначений рік відсутній у словнику часу.

## Завдання

2. Ініціалізуйте множину років, які ви хочете відвідати.
3. Реалізуйте обробку виключень для років, яких немає в вашому словнику часу.

## Додаткове завдання

Додайте функціональність, яка дозволяє користувачеві додавати та видаляти події із словника часу.

```python
#Вам потрібно створити список, кожен елемент якого буде кортежем із роком та подією в цьому році.
#Кожен кортеж повинен містити рік та опис події.
event_list = [
    ('1925', 'Перша публічна демонстрація телебачення'),
    ('1969', 'Народження Інтернету'),
    ('1939', 'Початок Другої світової війни'),
]

#Створіть словник, у якому ключі будуть роками, а значення — подіями.
event_dict = {
    1925: 'Перша публічна демонстрація телебачення',
    1969: 'Народження Інтернету',
    1939: 'Початок Другої світової війни',
}

#Створіть множину років, які ви хочете відвідати.
event_set = {1969,1925,1939,2001}

#Ваш код повинен коректно обробляти ситуацію, коли зазначений рік відсутній у словнику часу.
for year in event_set:
    try:
        print(f"{year} - {event_dict[year]}")
    except:
        print(f"{year} - немає в словнику")
```

```python
#Додайте функціональність, яка дозволяє користувачеві додавати та видаляти події із словника часу.
from enum import Enum

class MenuOptions(int, Enum):
    ADD_EVENT = 1,
    DELETE_EVENT = 2,
    VIEW_EVENTS = 3,
    EXIT = 4,

menu = 0

while menu!=MenuOptions.EXIT:
    try:
        menu = MenuOptions(int(input("Налаштування словника часу:\n1 - Додати подію;\n2 - Видалити подію;\n3 - Подивитися події;\n4 - Вийти з програми.\nВведіть значення: ")))
    except:
        print("Некоректне значення. Спробуйте ще раз.")

    match menu:
        case MenuOptions.ADD_EVENT:
            year_to_add = int(input("Введіть рік події: "))
            event_to_add = input("Введіть подію: ")
            
            if year_to_add not in event_dict:
                event_dict[year_to_add] = event_to_add
                print("Подію додано до словника часу.")
            else:
                print("Така подія вже є в словнику часу.")    

        case MenuOptions.DELETE_EVENT:
            event_to_delete = int(input("Введіть рік події: "))
            try:
                event_dict.pop(event_to_delete)
                print("Подію видалено зі словника часу.")
            except:
                print("Такої події нема в словнику часу.")
            
        case MenuOptions.VIEW_EVENTS:
            for year,event in event_dict.items():
                print(f"{year} - {event}")
            
        case MenuOptions.EXIT:
            print("Завершення програми.")
```
